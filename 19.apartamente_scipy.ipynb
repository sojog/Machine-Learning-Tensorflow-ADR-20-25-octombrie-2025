```python
from scipy import stats
```

```python
result = stats.linregress(suprafete, preturi)
result
```

```python

plt.plot(suprafete, preturi)
plt.scatter(suprafete, preturi);


plt.plot(suprafete, suprafete * result.slope + result.intercept, c="red") 
```
