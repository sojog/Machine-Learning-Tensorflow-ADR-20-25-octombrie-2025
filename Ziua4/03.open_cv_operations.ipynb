```python
gray_cat_image = cv2.imread('cat.jpg', cv2.IMREAD_GRAYSCALE)
gray_cat_image
```

```python
plt.imshow(gray_cat_image,  cmap='gray')
```

```python
gray_cat_image.shape
```

```python
resized_image = cv2.resize(gray_cat_image, (int(991 / 3), int(740 / 3)))

plt.imshow(resized_image,  cmap='gray')

cv2.imwrite("cat_resized.jpg", resized_image)
```

```python
blurred_image = cv2.blur(gray_cat_image, (50, 50))
cv2.imwrite("blurred_image.jpg", blurred_image)
```

```python
left_sobel_filter = np.array( [[1, 0, -1], [2, 0, -2],[1, 0, -1]] )
left_sobel_filter
```

```python
left_kernel_image =  cv2.filter2D(gray_cat_image, -1, left_sobel_filter)
plt.imshow(left_kernel_image, cmap="gray")
```

```python
top_sobel_filter = np.array([[ 1, 2,  1],
       [ 0,  0,  0],
       [ -1,  -2, -1]])
top_sobel_filter
```

```python
top_sobel_image =  cv2.filter2D(gray_cat_image, -1, top_sobel_filter)
plt.imshow(top_sobel_image, cmap="gray")
```
