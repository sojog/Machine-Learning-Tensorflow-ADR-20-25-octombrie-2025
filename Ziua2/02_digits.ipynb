```python
x_train, x_test, y_train, y_test = train_test_split(digits_data.data, digits_data.target, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier()
model
```

```python
model.fit(x_train, y_train)
```

```python
predicted_values = model.predict(x_test)
predicted_values
```

```python
pozitia = 100
print("A prezis:", predicted_values[pozitia], " si valoarea reala:", y_test[pozitia])


plt.imshow(x_test[pozitia].reshape(8,8), cmap="gray");
```

```python
model.score(x_test, y_test)
```
